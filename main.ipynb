```python
# write the function for limit approaches to zero to compute the derivative of any function
from math import e


def derivative(f, x):
    eps = 0.000000001
    return (f(x + eps) - f(x)) / eps


# test the function with a simple function

# def f(x):
#     return x ** 2

def complexFunction(x): 
    return x ** 2 + 2 * x + 1

# write the function whose derivative is undefined at x = 0
def fun2(x):
    return 1 / x

def relu(x) :
    return max(0.0, x)

f = relu

val = 2.0

print(derivative(f, val))

x = torch.tensor(val, requires_grad=True)
y = f(x)
y.backward()
print(x.grad)
```
